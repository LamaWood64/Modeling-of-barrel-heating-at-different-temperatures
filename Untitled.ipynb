```python
import numpy as np
import matplotlib.pyplot as plt

# -------------------------
# PARAMETERS (physical)
# -------------------------

r_i = 0.0045
r_o = 0.0075
L_seg = 0.04
A_cs = np.pi*(r_o**2 - r_i**2)
V_seg = A_cs * L_seg
rho = 7920.0
m_seg = rho * V_seg
c = 502.0

A_conv = 2*np.pi*r_o*L_seg

J_nom = 50.0
h_nom = 25.0
T_amb_list = [-40, -20, 0, 20, 40]

n_shots = 51
inter_shot = 0.3
shots_per_mag = 17
reload_time = 2.5

Nmc = 500
J_frac = 0.20
h_frac = 0.30
m_frac = 0.10
dt = 0.05

# -------------------------
# helpers
# -------------------------
def build_shot_times(N_shots, inter_shot, shots_per_mag=17, reload_time=2.5):
    shot_times = []
    t = 0.0
    shots_done = 0
    while shots_done < N_shots:
        shots_in_mag = min(shots_per_mag, N_shots - shots_done)
        for i in range(shots_in_mag):
            shot_times.append(t)
            shots_done += 1
            if i < shots_in_mag - 1:
                t += inter_shot
        if shots_done < N_shots:
            t += reload_time
    return np.array(shot_times)

def simulate_on_grid_given_params(shot_times, t_grid, J_shot, m, c, h, A_conv, T_env):
    T = np.ones_like(t_grid) * T_env
    tau = (m * c) / (h * A_conv)
    dT_shot = J_shot / (m * c)
    shot_set = set(np.round(shot_times / (t_grid[1]-t_grid[0])).astype(int))

    for i in range(1, len(t_grid)):
        dTdt = - (T[i-1] - T_env) / tau
        T[i] = T[i-1] + dTdt * (t_grid[i] - t_grid[i-1])

        if i in shot_set:
            T[i] += dT_shot
    return T


# -------------------------
# Prepare shot schedule and common time grid
# -------------------------
shot_times = build_shot_times(n_shots, inter_shot, shots_per_mag=shots_per_mag, reload_time=reload_time)
t_end = shot_times[-1] + 20.0
t_grid = np.arange(0.0, t_end + dt/2, dt)

# -------------------------
# Monte-Carlo: run per ambient temperature
# -------------------------
rng = np.random.default_rng(42)

results = {}

for T_env in T_amb_list:
    all_T = np.zeros((Nmc, t_grid.size))
    for i in range(Nmc):
        J_i = max(1e-3, rng.normal(J_nom, J_frac*J_nom))
        h_i = max(1e-3, rng.normal(h_nom, h_frac*h_nom))
        m_i = max(1e-6, rng.normal(m_seg, m_frac*m_seg))
        Tsim = simulate_on_grid_given_params(shot_times, t_grid, J_i, m_i, c, h_i, A_conv, T_env)
        all_T[i,:] = Tsim
    p5 = np.percentile(all_T, 5, axis=0)
    p50 = np.percentile(all_T, 50, axis=0)
    p95 = np.percentile(all_T, 95, axis=0)
    results[T_env] = {'t': t_grid, 'p5': p5, 'p50': p50, 'p95': p95, 'all': all_T}

# -------------------------
# Plot: one panel per ambient temperature
# -------------------------
ncols = 3
nrows = int(np.ceil(len(T_amb_list)/ncols))
fig, axes = plt.subplots(nrows, ncols, figsize=(5*ncols, 3*nrows), squeeze=False)
axes = axes.ravel()

for ix, T_env in enumerate(T_amb_list):
    ax = axes[ix]
    d = results[T_env]
    t = d['t']
    ax.fill_between(t, d['p5'], d['p95'], color='orange', alpha=0.3, label='5–95% band')
    ax.plot(t, d['p50'], color='red', linewidth=2, label='Median')
    ax.axhline(100, color='gray', linestyle='--', linewidth=0.7, alpha=0.8)
    ax.axhline(150, color='brown', linestyle='--', linewidth=0.7, alpha=0.8)
    ax.set_title(f"T_amb = {T_env} °C")
    ax.set_xlabel("Time (s)")
    ax.set_ylabel("Temperature (°C)")
    ax.grid(True)
    ax.legend(loc='upper left')
for j in range(len(T_amb_list), len(axes)):
    fig.delaxes(axes[j])
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

r_i = 0.0045; r_o = 0.0075; L_seg = 0.04
A_cs = np.pi*(r_o**2 - r_i**2)
V_seg = A_cs * L_seg
rho = 7920.0
m_seg = rho * V_seg
c = 502.0
A_conv = 2*np.pi*r_o*L_seg

J_shot_nom = 50.0
h_nom = 25.0
T_env = 20.0

def build_shot_times(N_shots, inter_shot, shots_per_mag=17, reload_time=2.5):
    shot_times = []
    t = 0.0
    shots_done = 0
    while shots_done < N_shots:
        shots_in_mag = min(shots_per_mag, N_shots - shots_done)
        for i in range(shots_in_mag):
            shot_times.append(t)
            shots_done += 1
            if i < shots_in_mag - 1:
                t += inter_shot
        if shots_done < N_shots:
            t += reload_time
    return np.array(shot_times)

def simulate_on_grid(shot_times, dt, J_shot, m=m_seg, c=c, h=h_nom, A=A_conv, T_env=20.0):
    t_end = shot_times[-1] + 10.0
    t_grid = np.arange(0.0, t_end + dt/2, dt)
    T = np.ones_like(t_grid) * T_env
    tau = (m * c) / (h * A)
    dT_shot = J_shot / (m * c)
    shot_indices = set((np.round(shot_times / dt)).astype(int))
    for i in range(1, len(t_grid)):
        dTdt = - (T[i-1] - T_env) / tau
        T[i] = T[i-1] + dTdt * dt
        if i in shot_indices:
            T[i] += dT_shot
    return t_grid, T

def inter_from_rpm(rpm):
    if rpm <= 0: return 1.0
    return 60.0 / rpm

rpm_list = [120, 180, 300]
labels = {120:"120 rpm (0.5 s)", 180:"180 rpm (0.333 s)", 300:"300 rpm (0.2 s)"}

plt.figure(figsize=(8,4))
for rpm in rpm_list:
    dt = inter_from_rpm(rpm)
    shot_times = build_shot_times(1, dt)
    t, T = simulate_on_grid(shot_times, dt=0.01, J_shot=J_shot_nom, h=h_nom, m=m_seg)
    plt.plot(t, T, label=f"{labels[rpm]}")
plt.xlabel("Time (s)"); plt.ylabel("Temperature (°C)")
plt.title("Single-shot cooling (same impulse) — different nominal firing rates (time scale differs)")
plt.legend(); plt.grid(True); plt.tight_layout()
plt.show()

plt.figure(figsize=(10,6))
for rpm in rpm_list:
    dt = inter_from_rpm(rpm)
    for Nshots, style in [(17,'-'), (51,'--'), (170,':')]:
        shot_times = build_shot_times(Nshots, dt)
        t, T = simulate_on_grid(shot_times, dt=0.02, J_shot=J_shot_nom, h=h_nom, m=m_seg)
        plt.plot(t, T, linestyle=style, label=f"{Nshots} shots @ {labels[rpm]}")
plt.xlabel("Time (s)"); plt.ylabel("Temperature (°C)")
plt.title("Temperature growth for different RPMs and sequence lengths")
plt.legend(fontsize=8); plt.grid(True); plt.tight_layout()
plt.show()

import numpy.random as rnd
rpm_mc = 180
dt_mc = 0.05
inter_shot_mc = inter_from_rpm(rpm_mc)
shot_times_mc = build_shot_times(51, inter_shot_mc)
t_grid_mc = np.arange(0.0, shot_times_mc[-1] + 20.0, dt_mc)

Nmc = 500
J_sigma = 0.20 * J_shot_nom
h_sigma = 0.30 * h_nom
m_sigma = 0.10 * m_seg

all_T = np.zeros((Nmc, t_grid_mc.size))
rng = rnd.default_rng(123)
for i in range(Nmc):
    J_i = max(1e-3, rng.normal(J_shot_nom, J_sigma))
    h_i = max(1e-3, rng.normal(h_nom, h_sigma))
    m_i = max(1e-6, rng.normal(m_seg, m_sigma))
    T = np.ones_like(t_grid_mc) * T_env
    tau_i = (m_i * c) / (h_i * A_conv)
    dT_i = J_i / (m_i * c)
    shot_idx = set((np.round(shot_times_mc / dt_mc)).astype(int))
    for k in range(1, len(t_grid_mc)):
        T[k] = T[k-1] - (T[k-1] - T_env)/tau_i * dt_mc
        if k in shot_idx:
            T[k] += dT_i
    all_T[i,:] = T

p5 = np.percentile(all_T, 5, axis=0)
p50 = np.percentile(all_T, 50, axis=0)
p95 = np.percentile(all_T, 95, axis=0)

plt.figure(figsize=(8,5))
plt.fill_between(t_grid_mc, p5, p95, color='orange', alpha=0.35, label='5–95% band (MC)')
plt.plot(t_grid_mc, p50, color='red', linewidth=2, label='Median (MC)')
plt.xlabel("Time (s)"); plt.ylabel("Temperature (°C)")
plt.title(f"Monte-Carlo (51 shots) @ {rpm_mc} rpm (inter_shot={inter_shot_mc:.3f}s)")
plt.legend(); plt.grid(True); plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

J_shot = 50
m = 0.0358
c = 502
h = 25
A = 0.001885
E = 200e9
alpha = 16.5e-6
nu = 0.3
sigma_yield_values = [600e6, 700e6]

dT_shot = J_shot / (m * c)
tau = (m * c) / (h * A)
k_sigma = E * alpha / (1 - nu)

def simulate_shooting(n_shots, dt, T_env):
    times = [0]
    temps = [T_env]
    T = T_env
    t = 0
    for i in range(n_shots):

        T += dT_shot
        times.append(t)
        temps.append(T)

        for k in range(int(dt*50)):
            t += dt/50
            T = T_env + (T - T_env) * np.exp(-dt/50 / tau)
            times.append(t)
            temps.append(T)

    return np.array(times), np.array(temps)

plt.figure(figsize=(8,5))

for T_env in [-20, 20, 40]:
    times, temps = simulate_shooting(51, 0.3, T_env)
    stresses = k_sigma * temps
    plt.plot(times, stresses/1e6, label=f"T_env={T_env} °C")

for sigma_yield in sigma_yield_values:
    plt.axhline(sigma_yield/1e6, color='red', linestyle='--',
                label=f"σ_yield={sigma_yield/1e6:.0f} MPa")
    plt.axhline(0.5*sigma_yield/1e6, color='orange', linestyle=':',
                label=f"0.5 σ_yield={0.5*sigma_yield/1e6:.0f} MPa")

plt.xlabel("Time (s)")
plt.ylabel("Stress σ (MPa)")
plt.title("Thermal stresses during 51-shot burst\nat different ambient temperatures")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from math import ceil

# ---------------------------
# Physical / material params
# ---------------------------
r_i, r_o, L_seg = 0.0045, 0.0075, 0.04
A_conv = 2*np.pi*r_o*L_seg
rho = 7920.0
m_nom = rho * np.pi*(r_o**2 - r_i**2) * L_seg   # ~0.0358 kg
c = 502.0

E = 200e9
alpha = 16.5e-6
nu = 0.30

k_sigma_MPa_perK = (E * alpha / (1 - nu)) / 1e6

J_nom = 50.0
h_nom = 25.0
A = A_conv

shots_per_mag = 17
reload_time = 2.5
inter_shot_mean = 0.3

shot_counts = [51, 85, 170, 340]

Nmc = 500
dt_step = 0.05
J_frac = 0.30
h_frac = 0.40
m_frac = 0.10

T_amb_list = [20.0, 40.0]
sigma_yield_list = [600.0, 700.0]

rng = np.random.default_rng(12345)

# ---------------------------
# helper: build shot times with random intervals for one realization
# returns array of shot times
# ---------------------------
def build_shot_times_random(N_shots, inter_mean, shots_per_mag=17, reload_time=2.5):
    shot_times = []
    t = 0.0
    shots_done = 0
    while shots_done < N_shots:
        shots_in_mag = min(shots_per_mag, N_shots - shots_done)
        for i in range(shots_in_mag):
            shot_times.append(t)
            shots_done += 1
            if i < shots_in_mag - 1:
                interval = rng.uniform(0.8*inter_mean, 1.2*inter_mean)
                t += interval
        if shots_done < N_shots:
            t += reload_time
    return np.array(shot_times)

# ---------------------------
# compute global t_grid long enough to contain max scenario
# ---------------------------
def estimate_max_t_end(shot_counts, inter_mean, shots_per_mag=17, reload_time=2.5):
    max_time = 0.0
    for N in shot_counts:
        mags = ceil(N / shots_per_mag)
        total = N * 1.2 * inter_mean + (mags - 1) * reload_time + 30.0
        if total > max_time:
            max_time = total
    return max_time

t_end_global = estimate_max_t_end(shot_counts, inter_shot_mean, shots_per_mag, reload_time)
t_grid = np.arange(0.0, t_end_global + dt_step/2, dt_step)

# ---------------------------
# main Monte-Carlo loop
# ---------------------------
summary = {}

for Nshots in shot_counts:
    print(f"Running MC for Nshots = {Nshots} ...")
    summary[Nshots] = {}
    for T_amb in T_amb_list:
        all_sigma = np.zeros((Nmc, t_grid.size))
        for i in range(Nmc):
            J_i = max(1e-9, rng.normal(J_nom, J_frac * J_nom))
            h_i = max(1e-9, rng.normal(h_nom, h_frac * h_nom))
            m_i = max(1e-9, rng.normal(m_nom, m_frac * m_nom))
            tau_i = (m_i * c) / (h_i * A)   # s
            dT_i = J_i / (m_i * c)

            shot_times = build_shot_times_random(Nshots, inter_shot_mean, shots_per_mag, reload_time)
            shot_indices = np.round(shot_times / dt_step).astype(int)

            dT_array = np.zeros_like(t_grid)
            valid_idx_mask = (shot_indices >= 0) & (shot_indices < t_grid.size)
            if np.any(valid_idx_mask):
                np.add.at(dT_array, shot_indices[valid_idx_mask], dT_i)

            T = np.empty_like(t_grid)
            T[0] = T_amb + dT_array[0]
            for k in range(1, len(t_grid)):
                T[k] = T[k-1] - (T[k-1] - T_amb) / tau_i * dt_step + dT_array[k]

            sigma_t = k_sigma_MPa_perK * (T - T_amb)
            all_sigma[i, :] = sigma_t

        p5 = np.percentile(all_sigma, 5, axis=0)
        p50 = np.percentile(all_sigma, 50, axis=0)
        p95 = np.percentile(all_sigma, 95, axis=0)
        max_per_run = all_sigma.max(axis=1)
        summary[Nshots][T_amb] = {
            't_grid': t_grid,
            'p5': p5,
            'p50': p50,
            'p95': p95,
            'max_per_run': max_per_run
        }

# ---------------------------
# Produce plots (one figure per Nshots, comparing T_amb 20 vs 40)
# ---------------------------
for Nshots in shot_counts:
    plt.figure(figsize=(9,5))
    for T_amb, col in zip(T_amb_list, ['tab:orange','tab:green']):
        d = summary[Nshots][T_amb]
        plt.fill_between(d['t_grid'], d['p5'], d['p95'], color=col, alpha=0.18)
        plt.plot(d['t_grid'], d['p50'], color=col, linewidth=2, label=f"T_amb={T_amb}°C (median)")
    # yield lines (MPa)
    for sy in sigma_yield_list:
        plt.axhline(sy, color='red', linestyle='--', label=f"σ_yield={sy:.0f} MPa")
        plt.axhline(0.5*sy, color='brown', linestyle=':', label=f"0.5·σ_yield={0.5*sy:.0f} MPa")
    plt.xlabel("Time (s)"); plt.ylabel("σ (MPa)")
    plt.title(f"Monte-Carlo σ(t), Nshots={Nshots}, Nmc={Nmc}")
    # avoid duplicate legend entries
    handles, labels = plt.gca().get_legend_handles_labels()
    by_label = dict(zip(labels, handles))
    plt.legend(by_label.values(), by_label.keys(), fontsize=8)
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# ---------------------------
# Tabulate exceedance probabilities
# ---------------------------
print("\nProbability that max σ (any time during run) exceeds thresholds")
headers = ["Nshots","T_amb (°C)","P(>0.5σ_y=600 MPa)","P(>σ_y=600 MPa)","P(>0.5σ_y=700 MPa)","P(>σ_y=700 MPa)"]
print("{:<8s} {:<10s} {:<18s} {:<16s} {:<18s} {:<16s}".format(*headers))
for Nshots in shot_counts:
    for T_amb in T_amb_list:
        max_runs = summary[Nshots][T_amb]['max_per_run']
        p_gt_05_600 = np.mean(max_runs >= 0.5 * 600.0)
        p_gt_600 = np.mean(max_runs >= 600.0)
        p_gt_05_700 = np.mean(max_runs >= 0.5 * 700.0)
        p_gt_700 = np.mean(max_runs >= 700.0)
        print(f"{Nshots:<8d} {T_amb:<10.1f} {p_gt_05_600:<18.3f} {p_gt_600:<16.3f} {p_gt_05_700:<18.3f} {p_gt_700:<16.3f}")
```
